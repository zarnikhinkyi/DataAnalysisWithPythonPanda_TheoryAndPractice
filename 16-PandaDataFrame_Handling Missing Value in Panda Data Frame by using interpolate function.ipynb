```python
data=pd.read_csv('data_temp.csv')
data
```

```python

data.interpolate(method='index')
```

```python
# limit_direction='forward' or 'backward', limit=1 mean one consecutive NaN value
data.interpolate(limit_direction='both', limit=1)
```
