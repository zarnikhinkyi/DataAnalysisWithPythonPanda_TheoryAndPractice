```python
data = pd.read_csv('data_m.csv')
data
```

# Drop the rows where at least one value is missing

```python
data.dropna(axis=0, how='any')
```

# Drop the columns where at least one value is missing

```python
data = pd.read_csv('data_m.csv')
data
```

```python
data.dropna(axis=1, how='any')
```

# Drop the rows where all values are missing

```python
data
```

```python
data.dropna(axis='index', how='all')
```

# Drop the columns where all values are missing

```python
data.dropna(axis='columns', how='all')
```

# Keep only the rows with at least 2 non-NA values

```python
data.dropna(axis=0, thresh=2)
```

# Define in which columns to look for missing/NA values

```python
data.dropna(subset='Marks')
```

```python
data.dropna(subset=['Marks','Grades'])
```
