### Replace values given in 'to_replace' with 'value
##### Syntex
- data.replace(to_replace=None, value=None, inplace=False, limit=None,
regex=False, mehtod='pad')

```python
import pandas as pd
```

```python
data=pd.read_csv('data_#na.csv')
data
```

#### Replace 'Priyang' with 'Anil'

```python
## to_replace='old value', value='new value'
data.replace(to_replace='Priyang', value='Anil')
## similar=>>> data.replace('Priyang','Anil')
```

#### Replace Name 'Priyang' and 'Aadhya' with 'Anil'

```python
data.replace(['Priyang','Aadhya'],'Anil')
```

#### Replace 'NaN' with 40

```python
import numpy as np
```

```python
data.replace(np.nan, 40)
```

#### Replace Marks column NaN with 40 and Grades column NaN with FF

```python
data.replace({'Marks':{np.nan:40},'Grades':{np.nan:'FF'}})
```

#### Replace 'not available', np.nan,'-99' of Grades column to 'FF'

```python
data.replace({'Grades':['not avaliable',np.nan,'-99']},'FF',inplace=True)
```

```python
data
```

#### Replace Grades values
#### FF:Fail
#### AB:Very Good
#### AA:Excellent
#### BA:Good


```python
data.replace(['FF','AB','AA','BA'],['Fail','Very Good','Excellent','Good'])
```

#### Replace NaN values with ffill

#### Replace NaN value (First only) of Grades column with 'FF 
### Using at

```python
Access a single value for a row/column label pair.
```

```python
data
```

```python
## we have to pass label pair [0,'Grades'], 0 meas first only
data.at[0,'Grades']='FF'
data
```

#### Replace consecutive NaN values of Grades column with 'FF'
### Using iloc

```python
iloc : Access a group of rows and columns by integer position(s).
```

```python
data=pd.read_csv('data_#na.csv')
data
```
