## 4. Insert the Column
- insert new column by using panda series with column position

```python
import pandas as pd

dict={'Name':['King','Queen','Pricess','Prince','Diana','William','Georg'],
      'Marks':[90,89,80,88,90,98,88]}
df1=pd.DataFrame(dict)
df1
```

## 6. Using Assign
- assign method will create new column with assign to new data frame

```python
import pandas as pd

dict={'Name':['King','Queen','Pricess','Prince','Diana','William','Georg'],
      'Marks':[90,89,80,88,90,98,88]}
df1=pd.DataFrame(dict)
df1
```

```python
series1=pd.Series(['AA','AB','BB','AB','AA','AA','AB'])
series2=pd.Series([11,22,33,44,55,66,77])
```

```python
## Create new column with Assign method >> assign to new df
df2=df1.assign(Grades=series1, Temp=series2)
```

```python
df2
```

```python
df1
```

```python
# add new column by using assign method . new column can be added by using old column value
df1.assign(Marks2=df1['Marks']+2)
```
