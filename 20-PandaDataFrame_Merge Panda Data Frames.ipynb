```python
import pandas as pd
```

```python
data1={'Key':['KO','K1','K2','K3','K4'],
      'Name':['Priyang','Aadhya','Vedant','Parshv','Krisha']}
data2={'Key':['KO','K1','K2','K3','K4'],
      'Marks':[98,99,89,85,87]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
# pass parameter is label
pd.merge(left=df1, right=df2, on='Key')
```

```python
##similar
pd.merge(df1,df2,on='Key')
```

```python
# similar
pd.merge(df1,df2)
```

# Two or more key parameters

```python
import pandas as pd
data1 = {'key1':['K0','k1','k2','k3','k4'],
         'key2':['K0','k1','k0','k1','k2'],
        'Name':['Priyang','Aadhya','Vedant','Parshv','krisha']}
data2={'key1':['K0','k1','k2','k3','k4'],
       'key2':['K0','k1','k0','k1','k2'],
      'Marks':[98,99,89,85,87]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
# pass parameter is list
pd.merge(df1,df2,on=['key1','key2'])
```

# parameter => how

```python
data1={'Key':['KO','K1','K2','K5','K6'],
      'Name':['Priyang','Aadhya','Vedant','Parshv','Krisha']}
data2={'Key':['KO','K1','K2','K3','K4'],
      'Marks':[98,99,89,85,87]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
# how='inner' means inner join
pd.merge(df1,df2,on='Key', how='inner')
```

```python
# how='left' means left join
pd.merge(df1,df2,on='Key', how='left')
```

```python
# how='right' means right join
pd.merge(df1,df2,on='Key', how='right')
```

```python
# how='right' means right join
pd.merge(df1,df2,on='Key', how='outer')
```

```python
# parameter =>indicator=True (boolean or string which wanted)
pd.merge(df1,df2,on='Key', how='outer', indicator=True)
```

```python
pd.merge(df1,df2,on='Key', how='outer', indicator="Sources Info")
```
